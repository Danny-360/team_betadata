```python
# Replace zeroes in features that can't be zero with NaN
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)
```

```python
df["BloodPressure"] = df["BloodPressure"].fillna(df["BloodPressure"].median())
df["Insulin"] = df["Insulin"].fillna(df["Insulin"].median())
```

```python
df["Glucose"] = df["Glucose"].fillna(df["Glucose"].median())
```

```python
df["SkinThickness"] = df["SkinThickness"].fillna(df["SkinThickness"].median())
df["BMI"] = df["BMI"].fillna(df["BMI"].median())
```

```python
# Define age bins and labels
bins = [0, 19, 29, 39, 49, 59, 69, 79, 89]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']
```

```python
# Create an age group column
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df.head(5)
```

```python
# Count number of diabetic cases (Outcome == 1) per age group
diabetes_by_age = df[df['Outcome'] == 1].groupby('AgeGroup',observed=True).size().reset_index(name='DiabetesCount')

print(diabetes_by_age)
```

```python
df["Outcome"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=diabetes_by_age, x='AgeGroup', y='DiabetesCount', palette='Set2')
plt.title('Number of Diabetic Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Diabetic Cases')
plt.tight_layout()
plt.show()
```

```python
# Define bins and labels for BloodPressure
bp_bins = [0, 59, 69, 79, 89, 99, 109, 200]  # adjust max as needed
bp_labels = ['<60', '60-69', '70-79', '80-89', '90-99', '100-109', '110+']
```

```python
# Define BMI bins (in kg/m²) and corresponding labels
bmi_bins = [0, 18.5, 24.9, 29.9, 34.9, 39.9, 100]  # Adjust max as needed
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese I', 'Obese II', 'Obese III']
```

```python
# Create a new column to categorize BMI
df['BMIGroup'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels, right=False)
```

```python
# View the new column
df[['BMI', 'BMIGroup']].head()
```

```python
# Filter diabetic cases and group by BMI category
bmi_diabetes_count = df[df['Outcome'] == 1].groupby('BMIGroup',observed=True).size().reset_index(name='DiabetesCount')

# View result
print(bmi_diabetes_count)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=bmi_diabetes_count, x='BMIGroup', y='DiabetesCount',color='pink')
plt.title('Number of Diabetic Cases by BMIGroup')
plt.xlabel('BMIGroup')
plt.ylabel('Number of Diabetic Cases')
plt.tight_layout()
plt.show()
```

```python
df['BMIGroup'].value_counts().plot(kind='pie', figsize=(10,7),autopct=('%1.1f%%'))
plt.title("BMIGroup Distribution")
plt.show()         
```

```python
# Define bins and labels
glucose_bins = [0, 70, 100, 126, df['Glucose'].max() + 1]
glucose_labels = ['Low', 'Normal', 'Prediabetic', 'Diabetic']
```

```python
# Create new column for Glucose group
df['GlucoseGroup'] = pd.cut(df['Glucose'], bins=glucose_bins, labels=glucose_labels, right=False)
```

```python
df[['Glucose', 'GlucoseGroup']].head()
```

```python
glucose_diabetes_count = df[df['Outcome'] == 1].groupby('GlucoseGroup', observed=True).size().reset_index(name='DiabetesCount')
print(glucose_diabetes_count)
```

```python
# Define bins and labels (you can adjust based on the data distribution)
skin_bins = [0, 10, 20, 30, 40, 50, df['SkinThickness'].max() + 1]
skin_labels = ['<10mm', '10-19mm', '20-29mm', '30-39mm', '40-49mm', '50mm+']
```

```python
# Create a new column for SkinThickness group
df['SkinGroup'] = pd.cut(df['SkinThickness'], bins=skin_bins, labels=skin_labels, right=False)
```

```python
df[['SkinThickness', 'SkinGroup']].head()
```

```python
skin_diabetes_count = df[df['Outcome'] == 1].groupby('SkinGroup',observed=True).size().reset_index(name='DiabetesCount')
print(skin_diabetes_count)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=skin_diabetes_count, x='SkinGroup', y='DiabetesCount', hue= "SkinGroup")
plt.title('Diabetes Count by Skin Thickness Group')
plt.xlabel('Skin Thickness Group (mm)')
plt.ylabel('Number of Diabetic Cases')
plt.tight_layout()
plt.show()
```

```python
#SkinThickness by Agegroup
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='AgeGroup', y='SkinThickness', color='Green', errorbar=None)
plt.title('Average SkinThickness by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Skin Thickness (mm)')
plt.tight_layout()
plt.show()
```

```python
#Variation of Skin thickness in diabetic and non-diabetic ptients across age groups
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='AgeGroup', y='SkinThickness', hue='Outcome', palette='Set2')
plt.title('SkinThickness by Age Group and Diabetes Outcome')
plt.xlabel('Age Group')
plt.ylabel('Skin Thickness (mm)')
plt.legend(title='Diabetes')
plt.tight_layout()
plt.show()
```
